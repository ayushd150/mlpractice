```python
X=df.drop('EngagementLevel', axis=1)
y=df['EngagementLevel']
X_train,X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
numeric_X = X.select_dtypes(include=["int64", "float64"])

# ---------------------------
# Compute correlation matrix
# ---------------------------
corr_matrix = numeric_X.corr()
print(corr_matrix)

# ---------------------------
# Extract upper triangle without diagonal
# ---------------------------
mask = np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
upper_triangle = corr_matrix.where(mask)

# ---------------------------
# Highest positive correlation
# ---------------------------
highest_positive_corr = upper_triangle.max().max()
print("Highest Positive Correlation Value:", highest_positive_corr)

# ---------------------------
# Total null values in entire dataset
# ---------------------------
total_nulls = df.isna().sum().sum()
print("Total null values:", total_nulls)
```

Rules for Imputing the missing(NaN) or Unknown values:
Calculating statistical values (such as mean, median, mode) for each column in the training dataset.

Applying these calculated statistical values to replace missing (NaN) and unknown values in both the training and test datasets.

Ensure that the calculation of statistical values excludes any rows containing missing or unknown values.

Replace Unknown values in the "Age" feature with the Mean value in that.

Replace Unknown values in the "Location" feature with the constant value "Other".

Replace Unknown values in the "GameDifficulty" feature with the Most Frequent value in that.

Replace Unknown values in the "InGamePurchases" feature with the constant value '0'.

Write the answers related to the above imputation in below questions respectively.

Write the sum of transformed(imputed) "Age" column of the test dataset . [NAT] (upto 2 digits after decimal points)

```python
# Mean for 'Age'
mean_age = X_train['Age'].mean()

# Mode for 'GameDifficulty'
mode_gamedifficulty = X_train['GameDifficulty'].mode()[0] # XXX = mode_gamedifficulty

# Imputation function (to be applied to both train and test)
def impute_data(X_data, mean_age, mode_gamedifficulty):
    # Age: Replace NaN/Unknown with Mean (train set)
    X_data['Age'] = X_data['Age'].fillna(mean_age)

    # Location: Replace NaN/Unknown with 'Other'
    X_data['Location'] = X_data['Location'].fillna('Other')

    # GameDifficulty: Replace NaN/Unknown with Most Frequent (train set)
    X_data['GameDifficulty'] = X_data['GameDifficulty'].fillna(mode_gamedifficulty)

    # InGamePurchases: Replace NaN/Unknown with '0'
    X_data['InGamePurchases'] = X_data['InGamePurchases'].fillna(0.0)

    return X_data

# Apply imputation to both X_train and X_test
X_train_imputed = impute_data(X_train.copy(), mean_age, mode_gamedifficulty)
X_test_imputed = impute_data(X_test.copy(), mean_age, mode_gamedifficulty)

# --- Final Calculations ---

# Sum of transformed(imputed) "Age" column of the test dataset
sum_age_test = X_test_imputed['Age'].sum()
print(sum_age_test)
# Value count of 'XXX' in "GameDifficulty" column of test dataset after imputation
count_gamedifficulty_test = X_test_imputed['GameDifficulty'].value_counts().get(mode_gamedifficulty, 0)
print(count_gamedifficulty_test)
```
