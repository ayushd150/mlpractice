```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('GA_5_dataset.csv')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Lasso, Ridge,RidgeCV, SGDRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score, cross_validate,KFold 
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.preprocessing import PolynomialFeatures,StandardScaler,OneHotEncoder,OrdinalEncoder
from sklearn.feature_selection import RFE
```

```python
X = df.drop('Credit_Limit', axis=1)
y = df['Credit_Limit']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression(fit_intercept=False)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
abs(lr.coef_).argmax()
```

```python
ridge = Ridge(solver='sag', tol=0.0005, random_state=42)
ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
ridge.intercept_
```

```python
lasso= Lasso(alpha=100, random_state=42)
lasso.fit(X_train, y_train)
lasso.score(X_test, y_test)
```

```python
((lasso.coef_>=-1)&(lasso.coef_<=1)).sum()
```

```python
# Import necessary libraries
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Create and train the model
knn = KNeighborsRegressor(n_neighbors=10, p=1)
knn.fit(X_train, y_train)

# Predict on test data
y_pred_knn = knn.predict(X_test)

# Compute RMSE
rmse_knn = np.sqrt(mean_squared_error(y_test, y_pred_knn))
print("KNeighbors RMSE:", round(rmse_knn, 3))
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(max_depth=10, min_samples_split=6, min_samples_leaf=6, random_state=42)
dtr.fit(X_train, y_train)
rmse = np.sqrt(mean_squared_error(y_test, dtr.predict(X_test)))
print(rmse)
```

```python
# Import necessary libraries
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score

# Define the model
ada = AdaBoostRegressor(random_state=42)

# Define parameter grid
param_grid = {
    'n_estimators': [10, 50, 100, 200, 500],
    'learning_rate': [0.1, 0.5, 1, 2]
}

# Perform GridSearchCV
grid = GridSearchCV(estimator=ada, param_grid=param_grid, cv=4)
grid.fit(X_train, y_train)

# Get best model
best_ada = grid.best_estimator_
print("Best Parameters:", grid.best_params_)

# Predict on test data
y_pred_ada = best_ada.predict(X_test)

# Compute RÂ² score
r2 = r2_score(y_test, y_pred_ada)
print("R2 Score on Test Data:", round(r2, 3))
```
